## Aggregation

```python
# aggregate to get the count of the number of items ordered per department 
# use the order_products data set
# reset the index and rename the columns after you've aggregated the data
ndepartment = orderProducts[["department","product_id"]].groupby(["department"]).agg("count")
ndepartment = ndepartment.reset_index()
ndepartment.columns = ["department","CountProduct_id"]
ndepartment
```

```python
# aggregate to get the count of the number of items ordered per aisle
# use the order_products data set
# reset the index and rename the columns after you've aggregated the data
naisle = orderProducts[["aisle","product_id"]].groupby(["aisle"]).agg("count")
naisle = naisle.reset_index()
naisle.columns = ["aisle","CountProduct_id"]
naisle
```

```python
# aggregate to get the count of the number of items ordered per day of week
# use the order_products data set
# reset the index and rename the columns after you've aggregated the data
norder_dow = orderProducts[["order_dow","product_id"]].groupby(["order_dow"]).agg("count")
norder_dow = norder_dow.reset_index()
norder_dow.columns = ["order_dow","CountProduct_id"]
norder_dow
# To interpret the data remember that 0 = Monday, 1 = Tuesday, 2 = Wednesday, etc
```

```python
# aggregate the orders data set to get the mean number of items ordered per order number
# reset the index and rename the columns after you've aggregated the data
norder_number = orders[["order_number","num_items"]].groupby(["order_number"]).agg("mean")
norder_number = norder_number.reset_index()
norder_number.columns = ["order_number","MeanNum_items"]
norder_number
```

```python
# aggregate the orders data set to get the median number of items ordered per order number
# reset the index and rename the columns after you've aggregated the data
norder_number_med = orders[["order_number","num_items"]].groupby(["order_number"]).agg("median")
norder_number_med = norder_number_med.reset_index()
norder_number_med.columns = ["order_number","MedianNum_items"]
norder_number_med
```

```python
# save the aggregated data set with the count of the number of items ordered per department to file
ndepartment.to_csv('agg.csv')
```

# Part B: Subsetting/Merging

```python
# example data sets
fruit_weights = pd.DataFrame({'fruit':['apples','bananas','apricots'],'weight_lbs':[0.50,.33,.10]})
fruit_prices = pd.DataFrame({'fruit':['oranges','bananas','apples'],'price_lbs':[1.39,0.58,2.99]}) 
```

## Subsetting

```python
# select only rows of fruit weights with weights less than .4 lbs
sub_weights = fruit_weights.loc[fruit_weights["weight_lbs"]<.4]
sub_weights
```

```python
# select only rows of fruit prices with prices higher than $1.00/lb
sub_prices = fruit_prices.loc[fruit_prices['price_lbs']>1.00]
sub_prices
```

```python
# select only rows of fruit prices with apples
sub_apple = fruit_prices.loc[fruit_prices['fruit']=="apple"]
sub_apple
```

## Merging

```python
# merge fruit weights and fruit prices. 
# your final data set should have *all* fruits (even those with missing values)
# get the number of rows and print the first few rows
weights_prices = pd.merge(fruit_weights,fruit_prices,how="outer",on=["fruit"],suffixes=["_weights","_prices"])
weights_prices
weights_prices.shape[0]
weights_prices.head()
```

```python
# merge fruit weights and fruit prices. 
# your final data set should *only* have fruits for which there is a weight and a price
# get the number of rows and print the first few rows
weights_prices2 = pd.merge(fruit_weights,fruit_prices,how="inner",on=["fruit"],suffixes=["_weights","_prices"])
weights_prices2
weights_prices2.shape[0]
weights_prices2.head()
```

```python
# merge organic and conventional data sets with how = "inner"
# get the number of rows and print the first few rows
org_conv = pd.merge(organic,conventional,how="inner",on=["aisle_id"],suffixes=["_org","_conv"])
org_conv
org_conv.shape[0]
org_conv.head()
```

```python
# merge organic and conventional data sets with how = "outer"
# get the number of rows and print the first few rows
org_conv2 = pd.merge(organic,conventional,how="outer",on=["aisle_id"],suffixes=["_org","_conv"])
org_conv2
org_conv2.shape[0]
org_conv2.head()
```

```python
# merge organic and conventional data sets with how = "left"
# get the number of rows and print the first few rows
org_conv3 = pd.merge(organic,conventional,how="left",on=["aisle_id"],suffixes=["_org","_conv"])
org_conv3
org_conv3.shape[0]
org_conv3.head()
```

```python
# merge organic and conventional data sets with how = "right"
# get the number of rows and print the first few rows
org_conv4 = pd.merge(organic,conventional,how="right",on=["aisle_id"],suffixes=["_org","_conv"])
org_conv4
org_conv4.shape[0]
org_conv4.head()
```

Questions:
- Why shouldn't you merge on "num_items"?
- How do the number of rows differ among the different merges?
- Why do the merges result in different data sets?
- Which merge is the appropriate choice? Why?

Answers:
- You shouldn't merge on "num_items" because there can be null or repeated values for "num_items".
- The number of rows differ slightly for each merge.  For example, the inner merge has 7 less rows than the outer merge.
- The merge result in different data sets because there are specific rows that are included and exluded depending on the type of merge.
- The outer merge is the appropriate choice.

# Part C: Reshaping/Calculating

## Reshaping

```python
# example data sets
fruit_orders = pd.DataFrame({'fruit':['apples','bananas','apricots'],'order1':[4,0,5],'order2':[1,3,0],'order3':[6,5,4]})
fruit_inventory = pd.DataFrame({'fruit':['apples','apples','oranges','oranges'],'store':['safeway','giant','safeway','giant'],'inventory':[100,75,62,59]})

```

```python
# convert fruit orders to long format
fruitLong = pd.melt(fruit_orders, id_vars="fruit",value_vars=["order1","order2","order3"],var_name="order",value_name="amount")
fruitLong
```

```python
# convert fruit inventory to wide format
fruitWide = fruit_inventory.pivot(index="fruit",columns="store",values="inventory")
fruitWide
```

```python
# convert merged conventional/organic data set how="outer" from Part B to long format
convert = pd.melt(org_conv2,id_vars="aisle_id",value_vars=["num_items","aisle"],var_name="number/aisle",value_name="number/aisle value")
convert
```

## Calculating

```python
# add a column with the price per item of fruit (use the merged data set Part B how = "outer")
# HINT: price per item = price/lbs * lbs
# print the first few rows
nweights_prices=weights_prices
nweights_prices["price_per_item"]=nweights_prices[["price_lbs","weight_lbs"]].apply(np.product,1)
nweights_prices.head()
```

```python
# add a column with the total number of items ordered (use fruit_orders data set)
# HINT: sum number of items ordered in order1, order2, order3
# print the first few rows
nfruit_orders=fruit_orders
nfruit_orders["total_items"]=nfruit_orders[["order1","order2","order3"]].apply(np.sum,1)
nfruit_orders.head()
```

```python
# add a column with the weight per item in grams instead of lbs (use the merged data set Part B how = "outer")
# use the apply with a lambda function
# 1 pound = 454 grams
# HINT: weight grams = weight lbs * (454)
# print the first few rows
weights_prices1 = weights_prices
weights_prices1["weight_grams"]=weights_prices1[["weight_lbs"]].apply(lambda x:x*454,1)
weights_prices1.head()
```
