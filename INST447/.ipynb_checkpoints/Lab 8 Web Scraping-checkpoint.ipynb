# Part B: Beautiful Soup

## Practice using beautiful soup

```python
# example html
html = """<!DOCTYPE html>
            <html>
            <head>
            <title>Page Title</title>
            </head>
            <body>

            <h1>This is a Heading</h1>
            <p>This is first paragraph.</p>
            <p>This is second paragraph.</p>
            <p>This is third paragraph.</p>

            </body>
            </html>
        """
```

```python
# convert the example html into beautiful soup
soup = BeautifulSoup(html, 'html.parser')

# print the beautifulsoup object (it should print the html)
print(soup)
```

```python
# use find to get the title of the examplt html website
title = soup.find("title")

# get the text within the title tags and print the title
for t in title:
    print(t)
```

```python
# use find_all to get the level 1 headings and print the content of within the header tags
heading = soup.find_all("h1")
for h in heading:
    for i in h:
        print(i)
```

```python
# use find_all to get the paragraphs and print the content of within the header tags
par = soup.find_all("p")
for p in par:
    for i in p:
        print(i)
```

## Using beautiful soup to get weather data

```python
# read in the text file you saved with the html from the darksky website
dstxt = open("darksky.txt","r")

#print this html from the darksky website
darksky.text
```

```python
# convert the html from the darksky website into a beautiful soup object
d = BeautifulSoup(dstxt, 'html.parser')

# print the beautifulsoup object created from the darksky website
print(d)
```

```python
# get and save the beautifulsoup elements that have the predicted minimum temperatures for each day
# HINT: use element inspector on the darksky website to notice where in the HTML the minimum temperature is stored
# HINT: look for a specific class of span elements
mintemp = d.find_all('span',{'class':'minTemp'})
# print the elements that contain the minimum temperatures for each day
print(mintemp)
```

```python
# get and save the beautifulsoup elements that have the predicted maximum temperatures for each day
# HINT: use element inspector on the darksky website to notice where in the HTML the maximum temperature is stored
# HINT: look for a specific class of span elements
maxtemp = d.find_all('span',{'class':'maxTemp'})
# print the elements that contain the maximum temperatures for each day
print(maxtemp)
```

```python
# use regular expressions to save just the minimum temperatures from the content of each element containing the minimum temperature
# HINT: mintempsnum = [int(re.search(yourregexp,elem.text).group(1)) for elem in mintemps]
mintempsnum = [int(re.search('\d{2}',elem.text).group()) for elem in mintemp]
# print the list of numbers you extracted
print(mintempsnum)
```

```python
# use regular expressions to save just the maximum temperatures from the content of each element containing the maximum temperature
maxtempsnum = [int(re.search('\d{2}',elem.text).group()) for elem in maxtemp]
# print the list of numbers you extracted
print(maxtempsnum)
```

```python
# create a data frame with these minimum and maximum temperatures
# by storing your two lists of temperatures in a dictionary and using the function pd.DataFrame
temps = {'min': mintempsnum, 'max':maxtempsnum}
tempsnum = pd.DataFrame(temps)
# print the first rows of this data frame
tempsnum.head()
```

```python
# create a scatterplot plotting the relationship between minimum and maximum temperatures
plt.scatter(mintempsnum,maxtempsnum)
```
