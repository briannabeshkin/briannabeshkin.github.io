```python
import pandas as pd
import numpy as np
```

```python
# “The Instacart Online Grocery Shopping Dataset 2017”, Accessed from https://www.instacart.com/datasets/grocery-shopping-2017 on 3/3/2020
# The data shows the number of products customers ordered based on the hour of day and product department
f = pd.read_csv("InstacartOrdersByDepartment.csv")
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# barchart of alcohol orders vs. hours
asubset = f.loc[f["department"] == "alcohol"]
plt.bar(asubset["order_hour_of_day"],asubset["num_orders_hour"])
plt.xlabel("Hour of Day")
plt.ylabel("Number of Orders")
plt.title("Alcohol Orders per Hour")
# Description: number of orders peak in the afternoon
# Research Question: many more people ordered food in this department later in the day than 
# in the early morning
# Take-away message: more orders of alcohol are made later in the day meaning that Instacart
# should have products in this department available to purchase and deliver during these times
# Note: This graph only includes orders in the Alcohol department
```
