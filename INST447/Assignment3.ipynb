```python
import pandas as pd
```

```python
# Read in files
WardIncome = pd.read_csv("WardIncome.csv")
requests = pd.read_csv("311_City_Service_Requests_in_2014.csv")
```

# Part A: Aggregating
Create data set of just parking related service requests (by Ward)

Transform a smaller unit to a larger unit of analysis by grouping variables based on the variables you want to summarize and then summarizing the function. Use the function, Group By, to group the variables.

```python
agg_requests = requests[["WARD","SERVICECODEDESCRIPTION"]].groupby(["WARD"]).agg("count")
agg_requests = agg_requests.reset_index()
agg_requests.columns = ["Ward", "Requests"]
agg_requests
```

# Part B: Subsetting & Merging

Filtering rows of the data frame by using boolean expressions. Then, use a merge function to combine data sets.

```python
# Filter Rows
pe_requests = requests.loc[requests["SERVICECODEDESCRIPTION"]=="Parking Enforcement"]
pe_requests
```

```python
pmr_requests = requests.loc[requests["SERVICECODEDESCRIPTION"]=="Parking Meter Repair"]
pmr_requests
```

```python
ROSA_requests = requests.loc[requests["SERVICECODEDESCRIPTION"]=="Out of State Parking Violation (ROSA)"]
ROSA_requests
```

```python
rrp_requests = requests.loc[requests["SERVICECODEDESCRIPTION"]=="Resident Parking Permit"]
rrp_requests
```

```python
rrpv_requests = requests.loc[requests["SERVICECODEDESCRIPTION"]=="Residential Parking Permit Violation"]
rrpv_requests
```

```python
# Combine Data Sets
combined1 = pd.merge(pe_requests,pmr_requests,how="outer",on=["OBJECTID"])
combined2 = pd.merge(combined1,ROSA_requests,how="outer",on=["OBJECTID"])
combined3 = pd.merge(combined2,rrp_requests,how="outer",on=["OBJECTID"])
combined = pd.merge(combined3,rrpv_requests,how="outer",on=["OBJECTID"])
combined
```

```python
#Create data set of just parking related service requests. Summarize the number of these requests by Ward.
parking_requests = combined[["WARD","SERVICECODEDESCRIPTION"]].groupby(["WARD"]).agg("count")
parking_requests = parking_requests.reset_index()
parking_requests.columns = ["Ward", "Parking_Requests"]
parking_requests
```

```python
#Create a data set of just non-parking related service requests. Summarize number of these requests by Ward.
npe_requests = requests.loc[requests["SERVICECODEDESCRIPTION"]!="Parking Enforcement"]
npmr_requests = npe_requests.loc[requests["SERVICECODEDESCRIPTION"]!="Parking Meter Repair"]
nrosa_requests = npmr_requests.loc[requests["SERVICECODEDESCRIPTION"]!="Out of State Parking Violation (ROSA)"]
nrpp_requests = nrosa_requests.loc[requests["SERVICECODEDESCRIPTION"]!="Resident Parking Permit"]
nrppv_requests = nrpp_requests.loc[requests["SERVICECODEDESCRIPTION"]!="Residential Parking Permit Violation"]

c1 = pd.merge(npe_requests,pmr_requests,how="outer",on=["OBJECTID"])
c2 = pd.merge(c1,nrosa_requests,how="outer",on=["OBJECTID"])
c3 = pd.merge(c2,nrpp_requests,how="outer",on=["OBJECTID"])
c = pd.merge(c3,nrppv_requests,how="outer",on=["OBJECTID"])

nonparking_requests = c[["WARD","SERVICECODEDESCRIPTION"]].groupby(["WARD"]).agg("count")
nonparking_requests = nonparking_requests.reset_index()
nonparking_requests.columns = ["Ward", "NonParking_Requests"]
nonparking_requests
```

# Part C: Reshaping

Organize data so that each variable has its own column, each observation has its own row and each value has its own cell using wide or long format as well as computing new variables.


```python
#Merge data sets
all_requests = pd.merge(parking_requests,nonparking_requests,how="inner",on=["Ward"])
all_requests
```

```python
all_requests["Total_Requests"] = (all_requests["Parking_Requests"]+all_requests["NonParking_Requests"])
all_requests
```
